```python
# Import Library
import pandas as pd
```

```python
# Create the function
def rankall(outcome, num):
    # Read the csv into a dataframe
    df = pd.read_csv('/home/stephen/Week4_Assignment/outcome-of-care-measures.csv')
    # Create a list with all the state abbreviations
    states = ["AL", "AK", "AZ", "AR", "CA", "CO", "CT", "DC", "DE", "FL", "GA", 
              "HI", "ID", "IL", "IN", "IA", "KS", "KY", "LA", "ME", "MD", 
              "MA", "MI", "MN", "MS", "MO", "MT", "NE", "NV", "NH", "NJ", 
              "NM", "NY", "NC", "ND", "OH", "OK", "OR", "PA", "RI", "SC", 
              "SD", "TN", "TX", "UT", "VT", "VA", "WA", "WV", "WI", "WY"]
    # Create an empty dictionary to store the state, hospital name key, value pairs
    hospitals = dict()
    # Loop through all the states
    for x in states:
        # Pull out all the data for hospitals in a specific state & store it in a new dataframe
        df_state = df[df['State'] == x]
        # Sort the values in the column of a specific outcome in ascending order
        df_state.sort_values(by=[outcome], inplace=True)
        # Count the number of non-NaN values in the specific outcome column
        # This only works if the cells in the csv say 'NaN' not 'Not Available' or anything else
        n_obs = df_state[outcome].count()
        # Tell the function to set the hospital name value for the state key equal to 
        # the hospital in the top cell of 'df_state' when certain conditions are met
        if num == 'best' and n_obs > 0:
            hospitals[x] = df_state.iloc[0]['Hospital Name']
        # Tell the function to set the hospital name value for the 
        # state key equal to 'NA' when certain conditions are met
        elif num == 'best' and n_obs == 0:
            hospitals[x] = 'NA'
        # Set what the function does 'num' is equal to 'worst' & 'n_obs' is greater than zero
        elif num == 'worst' and n_obs > 0:
            # Find the maximum value for a specific outcome
            max_value = df_state[outcome].max()
            # Find the name/s of the hospital/s with the maximum value
            name = df_state.loc[df_state[outcome] == max_value, 'Hospital Name']
            # 'name' contains a lot of extra info besides just 
            # the hospital name, so to you need to clean it up
            string_name = str(name)
            split = string_name.split()
            n_words = len(split)
            del split[n_words - 1]
            del split[n_words - 2]
            del split[n_words - 3]
            del split[n_words - 4]
            del split[n_words - 5]
            del split[0]
            # Conver the list back to a string
            string_join = ' '
            string_join = string_join.join(split)
            hospitals[x] = string_join
        # Tell the function to set the hospital name value for the 
        # state key equal to 'NA' when certain conditions are met
        elif num == 'worst' and n_obs == 0:
            hospitals[x] = 'NA'
        # Set the value for the state key equal to 'NA' if 
        # 'num' is greater than the rankings available
        elif num > n_obs:
            hospitals[x] = 'NA'
        # Loop through every possible ranking
        for y in list(range(0, n_obs)):
            # Set the hospital name value, associated with a specific ranking, for the state key
            if num == y + 1:
                hospitals[x] = df_state.iloc[y]['Hospital Name']
    # Convert the 'hospitals' dictionary to a dataframe
    df_final = pd.DataFrame(list(hospitals.items()),columns = ['state','hospital'])
    # Print
    print(df_final)
```

```python
try:
    rankall('Lower Mortality Estimate - Hospital 30-Day Death (Mortality) Rates from Heart Attack', 
            'best')
except:
    print('invalid outcome or state')
```

```python
try:
    rankall('Lower Mortality Estimate - Hospital 30-Day Death (Mortality) Rates from Heart Attack', 
            'worst')
except:
    print('invalid outcome or state')
```
